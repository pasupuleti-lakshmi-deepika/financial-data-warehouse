```python
## Data Summary
import pandas as pd

transactions = pd.read_csv("customers.csv")

print(transactions.info())
print(transactions.describe())
```

```python
## Remove Duplicates
import pandas as pd

customers = pd.read_csv("customers.csv")

customers = customers.drop_duplicates()

print(customers)
```

```python
## Handle Missing Values
import pandas as pd

accounts = pd.read_csv("accounts.csv")

accounts.fillna(0, inplace=True)

print(accounts)
```

```python
## Convert Date Format
import pandas as pd

transactions = pd.read_csv("transactions.csv")

transactions["transaction_date"] = pd.to_datetime(
    transactions["transaction_date"]
)

print(transactions.dtypes)
```

```python
## Currency Conversionimport pandas as pd

transactions = pd.read_csv("transactions.csv")
rates = pd.read_csv("exchange_rates.csv")

merged = transactions.merge(rates, on="currency")

merged["amount_usd"] = merged["amount"] * merged["rate_to_usd"]

print(merged)
```

```python
## Customer Balance Report
import pandas as pd

customers = pd.read_csv("customers.csv")
accounts = pd.read_csv("accounts.csv")

report = customers.merge(accounts, on="customer_id")

print(report)
```

```python
## Transaction Category
import pandas as pd

transactions = pd.read_csv("transactions.csv")

transactions["Category"] = transactions["amount"].apply(
    lambda x: "High" if x > 10000 else "Normal"
)

print(transactions)
```

```python
## Monthly Revenue
import pandas as pd

transactions = pd.read_csv("transactions.csv")

transactions["transaction_date"] = pd.to_datetime(
    transactions["transaction_date"]
)

transactions["Month"] = transactions["transaction_date"].dt.month

revenue = transactions.groupby("Month")["amount"].sum()

print(revenue)
```

```python
## Merge Three Tables
import pandas as pd

customers = pd.read_csv("customers.csv")
accounts = pd.read_csv("accounts.csv")
transactions = pd.read_csv("transactions.csv")

df = customers.merge(accounts, on="customer_id")
df = df.merge(transactions, on="account_id")

print(df)
```

```python
## High Value Customers
import pandas as pd

accounts = pd.read_csv("accounts.csv")

high_value = accounts[accounts["balance"] > 50000]

print(high_value)
```

```python
## Fraud Detection
import pandas as pd

transactions = pd.read_csv("transactions.csv")

fraud = transactions[transactions["amount"] > 20000]

print(fraud)
```

```python
## KPI Generation
import pandas as pd

transactions = pd.read_csv("transactions.csv")

print("Total Transactions:", len(transactions))
print("Total Amount:", transactions["amount"].sum())
print("Average Amount:", transactions["amount"].mean())
print("Maximum Amount:", transactions["amount"].max())
```

```python
## ETL Pipeline
import pandas as pd

customers = pd.read_csv("customers.csv")
accounts = pd.read_csv("accounts.csv")
transactions = pd.read_csv("transactions.csv")

final = customers.merge(accounts, on="customer_id")
final = final.merge(transactions, on="account_id")

final.to_csv("final_financial_data.csv", index=False)

print("ETL Pipeline Completed Successfully")
```

```python
## Load Processed Data into SQLite
import sqlite3
import pandas as pd

conn = sqlite3.connect("financial_dw.db")

df = pd.read_csv("final_financial_data.csv")

df.to_sql("financial_transactions", conn, if_exists="replace", index=False)

print("Data Loaded Successfully")

conn.close()
```
